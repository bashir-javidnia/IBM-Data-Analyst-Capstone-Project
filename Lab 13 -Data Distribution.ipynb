### Step 5: Visualize Job Satisfaction (Focus on JobSat)


- Create a pie chart or KDE plot to visualize the distribution of `JobSat`.

- Provide an interpretation of the plot, highlighting key trends in job satisfaction.


```python
jobsat_counts = df['JobSat'].value_counts()

plt.pie(jobsat_counts, labels=jobsat_counts.index, autopct='%1.1f%%', startangle=140)
plt.title('Distribution of Job Satisfaction (JobSat)')
plt.show()
```

### Step 6: Programming Languages Analysis


- Compare the frequency of programming languages in `LanguageHaveWorkedWith` and `LanguageWantToWorkWith`.
  
- Visualize the overlap or differences using a Venn diagram or a grouped bar chart.


```python
# 1. Extract and count languages from 'LanguageHaveWorkedWith'
# We split the strings by ';' and use explode to count each language individualy
languages_worked = df['LanguageHaveWorkedWith'].str.split(';').explode().value_counts()

# 2. Extract and count languages from 'LanguageWantToWorkWith'
languages_wanted = df['LanguageWantToWorkWith'].str.split(';').explode().value_counts()

# 3. Combine the counts into a single DataFrame for comparison
comparison_df = pd.DataFrame({
    'Worked With': languages_worked,
    'Want to Work With': languages_wanted
}).fillna(0)

# 4. Sort by popularity and select the top 10 for better visualization
top_languages = comparison_df.sort_values(by='Worked With', ascending=False).head(10)

# 5. Create a grouped bar chart
top_languages.plot(kind='bar', figsize=(12, 6), width=0.8)

plt.title('Top 10 Programming Languages: Current usage vs. Future Interest')
plt.xlabel('Programming Language')
plt.ylabel('Number of Respondents')
plt.xticks(rotation=45)
plt.legend()
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()

plt.show()
```

### Step 7: Analyze Remote Work Trends


- Visualize the distribution of RemoteWork by region using a grouped bar chart or heatmap.


```python
remote_region_df = df[['Country', 'RemoteWork']].dropna()

top_countries = remote_region_df['Country'].value_counts().nlargest(5).index
filtered_df = remote_region_df[remote_region_df['Country'].isin(top_countries)]

filtered_df = filtered_df.groupby(['Country', 'RemoteWork']).size().reset_index(name='Count')

plt.figure(figsize=(14, 8))
sns.barplot(data=filtered_df, x='Count', y='RemoteWork', hue='Country')

plt.title('Remote Work by Country')
plt.xlabel('Number of Respondents')
plt.ylabel('')
plt.legend(title='Country', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()
```

### Step 8: Correlation between Job Satisfaction and Experience


- Analyze the correlation between overall job satisfaction (`JobSat`) and `YearsCodePro`.
  
- Calculate the Pearson or Spearman correlation coefficient.


```python
!pip install Scipy
```

```python

# 1. Preprocessing 'YearsCodePro' 
# Convert strings like 'Less than 1 year' to 0 and 'More than 50 years' to 51
df['YearsCodePro'] = pd.to_numeric(df['YearsCodePro'].replace({
    'Less than 1 year': '0', 
    'More than 50 years': '51'
}), errors='coerce')

# 2. Ensure JobSat is numeric
df['JobSat'] = pd.to_numeric(df['JobSat'], errors='coerce')

# 3. Drop rows with missing values in either column for a clean analysis
corr_data = df[['JobSat', 'YearsCodePro']].dropna()

# 4. Calculate Correlation Coefficients
# Spearman is often preferred for JobSat as it is ordinal/rank-based
spearman_corr = corr_data['JobSat'].corr(corr_data['YearsCodePro'], method='spearman')
pearson_corr = corr_data['JobSat'].corr(corr_data['YearsCodePro'], method='pearson')

print(f"Spearman Correlation: {spearman_corr:.4f}")
print(f"Pearson Correlation: {pearson_corr:.4f}")
```

### Step 9: Cross-tabulation Analysis (Employment vs. Education Level)


- Analyze the relationship between employment status (`Employment`) and education level (`EdLevel`).

- **Instruction**: Create a cross-tabulation using `pd.crosstab()` and visualize it with a stacked bar plot if possible.


```python
# 1. Create the cross-tabulation
# This counts the number of respondents for each combination of EdLevel and Employment
top_emp = df['Employment'].value_counts().nlargest(5).index
filtered_df = df[df['Employment'].isin(top_emp)]

employment_edu_ct = pd.crosstab(filtered_df['EdLevel'], filtered_df['Employment'])

# 2. Visualize with a stacked bar plot
# 'stacked=True' allows us to see the composition of employment status within each education level
employment_edu_ct.plot(kind='bar', stacked=True)

plt.title('Relationship between Education Level and Employment Status')
# Using 'ha=right' and rotation to ensure long education level labels remain readable
plt.xlabel('Education Level')
plt.ylabel('Number of Respondents')
plt.legend(title='Employment Status', bbox_to_anchor=(1, 1), loc='upper left')
plt.xticks(rotation=45, ha='right')
plt.show()
```
